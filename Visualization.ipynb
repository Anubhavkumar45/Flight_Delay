```python
def barlabels(bars,c='black'):
    for bar in bars:
        yval = round(bar.get_height(),2)
        plt.text(bar.get_x()+0.25,yval, yval,verticalalignment='bottom',color=c,fontweight='bold')
```

```python
plt.figure(figsize=(10,7))
r= df.groupby('Month')
arr1=r['ArrDel15'].mean()
arr1=np.array(arr1)
arr1=arr1[1:]
bars1=plt.bar(x=np.arange(len(arr1)),height=arr1,color='#461C7C',label='Mean')
plt.xlabel('Month of year')
plt.xticks(np.arange(len(arr1)),['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'])
plt.ylabel('Percent of Flights that Arrive Late')
plt.title('Variation due to month')
plt.legend()
barlabels(bars1,'red')
plt.show()
```

```python
plt.figure(figsize=(20,5))
r= df.groupby('DayofMonth')
arr1=r['ArrDel15'].mean()
arr1=np.array(arr1)
bars1=plt.bar(x=np.arange(len(arr1)),height=arr1,color='black',label='Mean')
plt.xlabel('DayOfMonth')
plt.xticks(np.arange(len(arr1)))
plt.title('DayOfMonth')
plt.legend()
plt.ylabel('Percent of Flights that Arrive Late')
barlabels(bars1,'red')
plt.show()
```

```python
plt.figure(figsize=(7,7))
r= df.groupby('DayOfWeek')
arr1=r['ArrDel15'].mean()
arr1=np.array(arr1)
#arr1=arr1[1:]
bars1=plt.bar(x=np.arange(len(arr1)),height=arr1,color='grey',label='Mean')
plt.xlabel('Days of Week')
plt.xticks(np.arange(len(arr1)),['MON','TUE','WED','THUR','FRI','SAT','SUN'])
plt.title('Variation due to month')
plt.ylabel('Percent of Flights that Arrive Late')
plt.legend()
barlabels(bars1)
plt.show()
```

```python
plt.figure(figsize=(20,10))
r= df.groupby('DepTimeBlk')
arr=r['ArrDel15'].mean()
arr=np.array(arr)
bars=plt.bar(x=np.arange(len(arr)),height=arr,color='purple',label='Mean')
plt.xlabel('DepTimeBlk')
plt.xticks(np.arange(len(arr)),r['ArrDel15'].count().index,rotation=90)
plt.title('DepTimeBlk')
plt.ylabel('Percent of Flights that Arrive Late')
plt.legend()
barlabels(bars)
plt.show()
```

```python
plt.figure(figsize=(20,10))
r= df.groupby('ArrTimeBlk')
arr=r['ArrDel15'].mean()
arr=np.array(arr)
bars=plt.bar(x=np.arange(len(arr)),height=arr,color='purple',label='Mean')
plt.xlabel('ArrTimeBlk')
plt.xticks(np.arange(len(arr)),r['ArrDel15'].count().index,rotation=90)
plt.title('ArrTimeBlk')
plt.legend()
plt.ylabel('Percent of Flights that Arrive Late')
barlabels(bars)
plt.show()
```

```python
plt.figure(figsize=(20,10))
r= df.groupby('Reporting_Airline')
arr=r['ArrDel15'].mean()
arr=np.array(arr)
bars=plt.bar(x=np.arange(len(arr)),height=arr,color='pink',label='Mean')
plt.xlabel('Reporting_Airline')
plt.xticks(np.arange(len(arr)),r['ArrDel15'].count().index,rotation=90)
plt.title('Reporting_Airline')
plt.ylabel('Percent of Flights that Arrive Late')
plt.legend()
barlabels(bars)
plt.show()
```
